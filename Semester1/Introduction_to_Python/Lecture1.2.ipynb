```python
# Simple Interest Calculator
"""
Formula: Simple Interest = (Principal * Rate * Time) / 100
"""
principal = float(input("Enter principal amount: "))
rate = float(input("Enter rate of interest: "))
time = float(input("Enter time in years: "))
simple_interest = (principal * rate * time) / 100
total_amount = principal + simple_interest
print("Principal Amount: ",principal)
print("Rate of Interest: ",rate,"%")
print("Time Period: ",time ,"years")
print("Simple Interest: " ,simple_interest)
print("Total Amount: ",total_amount)
```

```python
"""
Create a program that:
1. Takes temperature in Celsius
2. Converts to Fahrenheit
3. Tells if water freezes, boils, or is liquid at that temp
Formula: F = (C * 9/5) + 32
"""

celsius = float(input("Enter temperature in Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"\n{celsius}째C = {fahrenheit}째F")

is_freezing = celsius <= 0
is_boiling = celsius >= 100
is_liquid = celsius > 0 and celsius < 100
print("Water freezes: ",is_freezing)
print("Water boils: ",is_boiling)
print("Water is liquid: ",is_liquid)
```

```python
"""
Simple Shopping Cart
"""
# Get cart items
item1_name = input("Enter item 1 name: ")
item1_price = float(input("Enter item 1 price: "))
item1_qty = int(input("Enter item 1 quantity: "))
item2_name = input("\nEnter item 2 name: ")
item2_price = float(input("Enter item 2 price: "))
item2_qty = int(input("Enter item 2 quantity: "))
item3_name = input("\nEnter item 3 name: ")
item3_price = float(input("Enter item 3 price: "))
item3_qty = int(input("Enter item 3 quantity: "))
# Calculate subtotals
item1_total = item1_price * item1_qty
item2_total = item2_price * item2_qty
item3_total = item3_price * item3_qty

items_total = item1_total + item2_total + item3_total
# Apply discount if items_total > 1000
has_discount = items_total > 1000
discount_rate = 10 if has_discount else 0
discount_amount = (items_total * discount_rate) / 100
after_discount = items_total - discount_amount
# Calculate tax (18% GST)
tax_rate = 18
tax_amount = (after_discount * tax_rate) / 100
final_total = after_discount + tax_amount
# Display bill
print("\n" + "="*50)
print("INVOICE")
print("="*50)
print("Item: " ,item1_name , "Price: " , item1_price , "Qty: ", item1_qty, "Total: ", item1_total)
print("Item: ",item2_name, "Price: ", item2_price, "Qty: ", item2_qty, "Total: ", item2_total)
print("Item: ",item3_name, "Price: ", item3_price, "Qty: ", item3_qty, "Total: ", item3_total)
print("Discount: ", discount_amount)
print("Tax: ", tax_amount)
print("Total Bill:" , final_total)
```
