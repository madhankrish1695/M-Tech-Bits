```python
import numpy as np


grades = np.array([85, 92, 78, 95, 88, 76, 90, 82, 91, 87])
print("Original grades:")
print(grades)

print("\nMean (Average):")
print(np.mean(grades))

print("\nMedian:")
print(np.median(grades))

print("\nStandard Deviation:")
print(np.std(grades))

print("\nMaximum grade:")
print(np.max(grades))

print("\nMinimum grade:")
print(np.min(grades))

print("\nIndex of highest grade:")
print(np.argmax(grades))

print("\nIndex of lowest grade:")
print(np.argmin(grades))

print("\nSum of all grades:")
print(np.sum(grades))

print("\nGrades above 85:")
print(grades[grades > 85])

print("\nAdd 5 bonus points to all grades:")
bonus_grades = grades + 5
print(bonus_grades)

print("\nMultiply all grades by 1.1 (10% curve):")
curved_grades = grades * 1.1
print(curved_grades)

print("\nSort grades in ascending order:")
print(np.sort(grades))

print("\nSort grades in descending order:")
print(np.sort(grades)[::-1])

print("\nReshape into 5 rows, 2 columns:")
reshaped = grades.reshape(5, 2)
print(reshaped)

print("\nCreate array of zeros (10 elements):")
zeros = np.zeros(10)
print(zeros)

print("\nCreate array of ones (5 elements):")
ones = np.ones(5)
print(ones)

print("\nCreate array from 0 to 9:")
range_array = np.arange(10)
print(range_array)

print("\nCreate array with 5 evenly spaced values between 0 and 100:")
linspace_array = np.linspace(0, 100, 5)
print(linspace_array)

print("\nRandom grades (10 students, between 60-100):")
random_grades = np.random.randint(60, 101, 10)
print(random_grades)
```

```python
import numpy as np

print("Weekly Temperature Analysis\n")

temps_celsius = np.array([22, 25, 28, 30, 27, 24, 23])
days = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

print("Daily temperatures (°C):")
for i in range(len(days)):
    print(f"{days[i]}: {temps_celsius[i]}°C")

print(f"\nAverage temperature: {np.mean(temps_celsius):.1f}°C")
print(f"Highest temperature: {np.max(temps_celsius)}°C")
print(f"Lowest temperature: {np.min(temps_celsius)}°C")

temps_fahrenheit = (temps_celsius * 9/5) + 32
print("\nTemperatures in Fahrenheit:")
for i in range(len(days)):
    print(f"{days[i]}: {temps_fahrenheit[i]:.1f}°F")
```

```python
# Import matplotlib's pyplot module and give it the shortcut name 'plt'
import matplotlib.pyplot as plt

# Create a list of x-axis values
x = [1, 2, 3, 4]

# Create a list of y-axis values
y = [11, 22, 33, 44]

# Create a figure (canvas) and axes (plotting area) with size 5 inches by 5 inches
fig, ax = plt.subplots(figsize=(5,5))

# Draw a line connecting the points (1,11), (2,22), (3,33), (4,44)
ax.plot(x, y)

# Add title to the plot and labels to both axes
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# Display the plot in a window
plt.show()
```

IMDB Dataset Visualization

```python
df = pd.read_csv("IMDB-Movie-Data.csv")
fig = plt.figure(figsize=(10, 10))
plt.hist(df['Rating'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Ratings')
plt.grid(axis='y', alpha=0.3)

```

```python
genre_counts = df['Genre'].value_counts().head(10)
plt.barh(genre_counts.index, genre_counts.values, color='coral')
plt.xlabel('Count')
plt.title('Top 10 Genres')
```

```python
year_counts = df['Year'].value_counts().sort_index()
plt.plot(year_counts.index, year_counts.values, color='green', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.title('Movies Released Over Time')
```

```python
plt.scatter(df['Rating'], df['Revenue (Millions)'], alpha=0.5, color='orange', s=30)
plt.xlabel('Rating')
plt.ylabel('Revenue (Millions)')
plt.title('Rating vs Revenue')
```

```python
top_directors = df['Director'].value_counts().head(10)
plt.bar(range(len(top_directors)), top_directors.values, color='teal')
plt.xticks(range(len(top_directors)), top_directors.index, rotation=45, ha='right')
plt.ylabel('Number of Movies')
plt.title('Top 10 Most Prolific Directors')
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
sns.histplot(data=tips, x='total_bill', bins=20, color='skyblue')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', s=80)
plt.title('Tips vs Total Bill by Gender and Smoker Status')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
```

```python
sns.barplot(data=tips, x='day', y='tip')
plt.title('Average Tip by Day of Week')
plt.xlabel('Day')
plt.ylabel('Average Tip ($)')
```

```python
sns.boxplot(data=tips, x='time', y='tip', hue='sex', palette='Set2')
plt.title('Tips Distribution by Time and Gender')
plt.xlabel('Time')
plt.ylabel('Tip ($)')
```
